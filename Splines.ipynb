```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

```python
def weirdLine(x):
    return 2 + x * 2 + max(0, x - 5) * -4 + max(0, x - 10) * 10
```

```python
xs = np.sort(np.random.uniform(-5, 15, size = 1000))
ys = map(weirdLine, xs) + np.random.normal(0, 2, size = 1000) + 2
plt.scatter(xs, ys)
```

```python
from splines import SplineRegression
spl = SplineRegression()
spl.fit(xs, ys, [5,10])
fig, ax = plt.subplots()
ax.scatter(xs, ys, alpha = .3)
ax.plot(xs, spl.predict(xs), "--k")
```
